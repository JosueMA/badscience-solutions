## Individual agents

```python
ed.hdf.close()
ed = ExperimentData('FprAllAgents.hdf')

Gs = [10, 35, 60, 85]
figsize = (8.5, 2.5)
def save_path(param, G):
    return '/Users/mt/workspace/papers/sciencefunding/Figures/fpr-over-{}-allagents-G={}.pdf'.format(
        param, G
    )

for G in Gs:
    
    param = 'NPR'
    fpr_allpi(
        ed, G=G, param=param, 
        save_path=save_path(param, G),
        figsize=figsize
    )
    
    plt.figure()
    
    param = 'FPDR'
    fpr_allpi(
        ed, G=G, param=param, figsize=figsize, 
        save_path=save_path(param, G)
    )
```
